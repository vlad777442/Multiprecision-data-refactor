```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'packet_loss_data2048nosleep.csv'
packet_loss_data = pd.read_csv(file_path)
packet_loss_data['packet_loss_rate'] = packet_loss_data['packet_loss_rate'].apply(lambda x: x * 1000)

plt.figure(figsize=(8, 4))
plt.hist(packet_loss_data['packet_loss_rate'], bins=1000, color='skyblue', edgecolor='black')
plt.title('Histogram of Packet Loss 8192')
plt.xlabel('Packet Loss Rate')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.xlim(0, 10000)

plt.show()
```
